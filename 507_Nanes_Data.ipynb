```python
## Question 1: "Is there an association between marital status (married or not married) and education level (bachelor’s degree or higher vs. less than a bachelor’s degree)?"

## Variables: DMDMARTZ (marital status) and DMDEDUC2 (education level).

## Test: I'm using the Chi-square test to find the correlation instead of a pearson's or spearmen's test since these variables are categorical.


demographic_path = 'DEMO_L.XPT'
demo = pd.read_sas(demographic_path, format='xport')

# Recode Marital Status (DMDMARTZ) and Education Level (DMDEDUC2)
demo['Marital_Status'] = demo['DMDMARTZ'].apply(lambda x: 'Married' if x in [1, 2] else 'Not Married')
demo['Education_Level'] = demo['DMDEDUC2'].apply(lambda x: 'Bachelor’s or Higher' if x >= 4 else 'Less than Bachelor’s')

# Create a contingency table
contingency_table = pd.crosstab(demo['Marital_Status'], demo['Education_Level'])

print(contingency_table)
```

```python
# Perform the Chi-square test and show the P-value
chi2, p, dof, ex = stats.chi2_contingency(contingency_table)
contingency_table, chi2, p
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
```

```python
# Print Answer to Q1:
if p < 0.05:
    print("There is a significant association between marital status and education level.")
else:
    print("There is no significant association between marital status and education level.")
```

```python
# Question 2: "Is there a difference in the mean sedentary behavior time between those who are married and those who are not married?"

# Variables: DMDMARTZ (marital status, recoded) and PAD680 (sedentary behavior time, cleaned).

#Test: I'm using a T-test to determine the significance between the means of the variables
```

```python
# Load Measurement Data: Physical Activity

physical_activity_path = 'PAQ_L.XPT'
physical_activity = pd.read_sas(physical_activity_path, format='xport')
physical_activity
```

```python
# Clean the PAD680 variable in the physical_activity by removing values 7777, 9999, and null

physical_activity = physical_activity[physical_activity['PAD680'] != 7777]
physical_activity = physical_activity[physical_activity['PAD680'] != 9999]
physical_activity = physical_activity.dropna(subset=['PAD680'])
```

```python
# Merge the variables from two tables into one dataset

merged_data = pd.merge(demo[['SEQN', 'DMDMARTZ']], physical_activity[['SEQN', 'PAD680']], on='SEQN', how='inner')
merged_data
```

```python
# Create two groups from the variable DMDMARTZ for those who are married and those who are not
married_data = merged_data[merged_data['DMDMARTZ'] == 1]
not_married_data = merged_data[merged_data['DMDMARTZ'] == 2]
```

```python
# Perform a T-test between married, not married, and physical activity

t_statistic, p_value = stats.ttest_ind(married_data['PAD680'], not_married_data['PAD680'])

# Print results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Print Response to Q2
if p_value < 0.05:
    print("There is a significant difference in the mean sedentary behavior time between those who are married and those who are not married.")
if t_statistic > 0:
    print("Those who are married have a higher mean sedentary behavior time than those who are not married.")
else:
    print("Those who are not married have a higher mean sedentary behavior time than those who are married.")
```

```python
# Question 3: "How do age and marital status affect systolic blood pressure?"
# Variables: RIDAGEYR (age), DMDMARTZ (marital status, recoded), and BPXOSY3 (systolic blood pressure).
# Test: Using a 2-way ANOVA test since age and marital status are independent of each other but are needed to determine significance for blood pressure
```

```python
# Merge the three variables into one table

merged_data = pd.merge(demo[['SEQN', 'RIDAGEYR', 'DMDMARTZ']], bp[['SEQN', 'BPXOSY3']], on='SEQN', how='inner')
merged_data
```

```python
# Perform a 2-way ANOVA test on these variables

model = sm.formula.ols(formula='BPXOSY3 ~ C(RIDAGEYR) + C(DMDMARTZ)', data=merged_data).fit()
table = sm.stats.anova_lm(model, typ=2)
print(table)

# Response to Q3
if table['PR(>F)']['C(RIDAGEYR)'] < 0.05 and table['PR(>F)']['C(DMDMARTZ)'] < 0.05:
    print("Age and marital status have significant effect on systolic blood pressure.")
else:
    print("Age and marital status do not have significant effect systolic blood pressure.")
```

```python
# Question 5 (Creative Analysis): Is there a correlation with education level (bachelor's degree or higher) and general health condition?
# Variables: DMDEDUC2 (education level) and HUQ010 (general health condition)
# Test: A Chi-square test since both variables contain categorical data
```

```python
# Recode the data to indicate a degree higher than a bachelors and general health condition as fair or very good
demo['Education_Level'] = demo['DMDEDUC2'].apply(lambda x: 'Bachelor’s or Higher' if x >= 4 else 'Less than Bachelor’s')
access['General_Health_Condition'] = access['HUQ010'].apply(lambda x: 'Fair or Very Good' if x in [2, 4] else 'Other')
```

```python
# Perform a chi-square test between these two recoded variables
contingency_table = pd.crosstab(demo['Education_Level'], access['General_Health_Condition'])
chi2, p, dof, ex = stats.chi2_contingency(contingency_table)

# Print results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# Response to Q5
if p < 0.05:
    print("There is a significant correlation between education level and general health condition.")
else:
    print("There is no significant correlation between education level and general health condition.")
```
